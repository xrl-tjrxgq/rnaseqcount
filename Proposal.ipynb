## 一、项目意向
Project 13：rnaseqcount——RNA-seq Count Matrix 分析包
## 二、项目分析
### 2.1 生物学背景
在转录组学研究中，研究人员通过RNA-seq比较不同条件下（如疾病vs正常）的基因表达水平。首先从生物样本中提取RNA，反转录为cDNA并进行测序，产生包含数百万条短序列的FASTQ文件；然后将这些序列比对到参考基因组上，确定每条序列来自哪个基因；最后用计数工具统计每个基因被比对到的序列数量，就得到了一个行代表基因、列代表样本、值为整数计数的Count Matrix.
### 2.2 项目目标
给定 bulk RNA-seq count matrix，利用其系统地进行标准化、质量控制和差异表达分析。具体来说，即开发一个轻量级 Python 工具包 `rnaseqcount`，对bulk RNA-seq count matrix类数据实现：
| 功能 | 方法 |
|------|------|
| 数据读取 | count matrix + metadata |
| 标准化 | CPM/TPM/FPKM转换 |
| 过滤 | 低表达基因过滤 |
| 可视化 | PCA、相关性热图、火山图 |
| 差异表达 | 简化 t-test/Wilcoxon/negative binomial GLM 示例 |
### 2.3 包模块职责分配
| 文件 | 职责 |
|------|------|
| `io.py` | 读取 raw count matrix、sample metadata |
| `core.py` | 定义 `CountData` 类，存储 counts、metadata，作为统一数据容器 |
| `stats.py` | 低表达基因过滤、CPM/TPM/FPKM 标准化、PCA 计算、样本间相关性 |
| `algorithms.py` | 简化差异表达：t-test、Wilcoxon rank-sum、负二项 GLM 示例 |
| `visualize.py` | PCA 散点图、样本相关性热图、火山图 |
| `cli.py` | 命令行接口 |
### 2.4 数据来源
主要使用NCBI GEO数据库
